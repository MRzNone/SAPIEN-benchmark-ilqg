```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
a = []
b = []


fig = plt.figure()
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

for i in range(0,100):
    ax.clear()
    
    c = np.random.random()
    d = np.random.random()
    
    a.append(c)
    b.append(d)
    
    ax.plot(a,)
    ax.plot(b)
    
    fig.canvas.draw()
    fig.show()
```
